# Section 5 : Data Science
### I would like to chose a different question than I proposed in section 4 because I am more interested in answering the question do forwards score more goals than the league average of 3 goals per season. 
## Hypotheses:
### Null: The average goals by forwards is 3.
### Alternative: The average goals by forwards is not 3
## I plan on using a 1 sample t-test using python to answer this by isolating the goals column and then interpret the p-value to determine the significance at a 5% level

```python
from scipy.stats import ttest_1samp
```

```python
data = [15, 3, 5, 9, 4, 0]
from scipy.stats import ttest_1samp
ttest_1samp(data, popmean=0)
```

### The p-value is 0.039019  which tells us the probability of observing a sample mean when we assume the null hypothesis is true. In the context of my dataset it's the probability that the forwards average goals per season would be as far from 3 as I observed.
